```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def read_data(file):
    try:
        data=pd.read_csv(file)
        return data
    except FileNotFoundError as fl:
        print("file read error",fl)
```

```python
file="E:\Datasets\energydata_complete.csv"
data=read_data(file)
```

```python
data.head()
```

```python
data.info()
```

## histogram 

```python
data=data.drop('date',axis=1)
row=4
cols=7

col=data.columns
fig,ax=plt.subplots(nrows=row,ncols=cols,figsize=(16,4))
index=0
for i in range(row):
    for j in range(cols):
        sns.distplot(data[col[index]],ax=ax[i][j])
        index=index+1
plt.tight_layout()
```

## Boxplot

```python
row=4
cols=7

col=data.columns
fig,ax=plt.subplots(nrows=row,ncols=cols,figsize=(16,4))
index=0
for i in range(row):
    for j in range(cols):
        sns.boxplot(data[col[index]],ax=ax[i][j])
        index=index+1
plt.tight_layout()
```

## corelation

```python
fig,ax=plt.subplots(figsize=(16,10))
sns.heatmap(data.corr())
```
