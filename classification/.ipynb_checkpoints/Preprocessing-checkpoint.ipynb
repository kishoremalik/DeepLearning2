## Reading data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def read_data(file):
    try:
        data=pd.read_csv(file)
        return data
    except FileNotFoundError as fl:
        print("file read error",fl)
file="E:\Datasets\energydata_complete.csv"
data=read_data(file)
```

## checking out liers an manage

```python
data=data.drop('date',axis=1)
row=4
cols=7

col=data.columns
fig,ax=plt.subplots(nrows=row,ncols=cols,figsize=(16,4))
index=0
for i in range(row):
    for j in range(cols):
        sns.boxplot(data[col[index]],ax=ax[i][j])
        index=index+1
plt.tight_layout()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

data_final = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
data_final.shape
```

```python
row=4
cols=7

col=data_final.columns
fig,ax=plt.subplots(nrows=row,ncols=cols,figsize=(16,4))
index=0
for i in range(row):
    for j in range(cols):
        sns.boxplot(data_final[col[index]],ax=ax[i][j])
        index=index+1
plt.tight_layout()
```

```python
row=4
cols=7

col=data_final.columns
fig,ax=plt.subplots(nrows=row,ncols=cols,figsize=(16,4))
index=0
for i in range(row):
    for j in range(cols):
        sns.distplot(data_final[col[index]],ax=ax[i][j])
        index=index+1
plt.tight_layout()
```
