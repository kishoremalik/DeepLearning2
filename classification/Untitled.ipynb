```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('E:\\Datasets\\diabetes.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,ax=plt.subplots(figsize=(16,10))
sns.heatmap(data.corr(),annot=True,cmap='coolwarm')
```

```python
train_data=data[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]]
```

```python
target=data["Outcome"]
```

```python
from sklearn.preprocessing import StandardScaler
scl=StandardScaler()
scl.fit(train_data.values)
scl_data=scl.transform(train_data.values)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scl_data,target,test_size=0.2,random_state=2)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel="rbf")
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,pred))
```

```python
accuracy_score(y_test,pred)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
logpred=logmodel.predict(x_test)
```

```python
print(classification_report(y_test,logpred))
```

```python
accuracy_score(y_test,logpred)
```
